```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('../data/all_matches.csv')
df.fillna(0)
# print(dataset.columns)
```

```python
df.replace('MA Chidambaram Stadium, Chepauk','MA Chidambaram Stadium',inplace=True)
df.replace('MA Chidambaram Stadium, Chepauk, Chennai','MA Chidambaram Stadium',inplace=True)
df.replace('M.Chinnaswamy Stadium','M Chinnaswamy Stadium',inplace=True)
df.replace('Feroz Shah Kotla','Arun Jaitley Stadium',inplace=True)
df.replace('Wankhede Stadium','Wankhede Stadium, Mumbai',inplace=True)
df.replace('Punjab Cricket Association Stadium, Mohali, Mumbai','Punjab Cricket Association IS Bindra Stadium',inplace=True)
df.replace('Punjab Cricket Association IS Bindra Stadium, Mohali','Punjab Cricket Association IS Bindra Stadium',inplace=True)
df.replace('Rajiv Gandhi International Stadium, Uppal','Rajiv Gandhi International Stadium',inplace=True)
df.replace('Subrata Roy Sahara Stadium','Maharashtra Cricket Association Stadium',inplace=True)
df.replace('M Chinnaswamy Stadium','MA Chidambaram Stadium',inplace=True)

print(df['venue'].unique())
```

```python
main = pd.DataFrame(columns= ['match_id','venue','innings','batting_team','bowling_team','runs_6overs','wickets'])
df_test1 = pd.read_csv('../inputFile.csv')

df_test1.replace('MA Chidambaram Stadium, Chepauk','MA Chidambaram Stadium',inplace=True)
df_test1.replace('MA Chidambaram Stadium, Chepauk, Chennai','MA Chidambaram Stadium',inplace=True)
df_test1.replace('M.Chinnaswamy Stadium','M Chinnaswamy Stadium',inplace=True)
df_test1.replace('Feroz Shah Kotla','Arun Jaitley Stadium',inplace=True)
df_test1.replace('Wankhede Stadium','Wankhede Stadium, Mumbai',inplace=True)
df_test1.replace('Punjab Cricket Association Stadium, Mohali, Mumbai','Punjab Cricket Association IS Bindra Stadium',inplace=True)
df_test1.replace('Punjab Cricket Association IS Bindra Stadium, Mohali','Punjab Cricket Association IS Bindra Stadium',inplace=True)
df_test1.replace('Rajiv Gandhi International Stadium, Uppal','Rajiv Gandhi International Stadium',inplace=True)
df_test1.replace('Subrata Roy Sahara Stadium','Maharashtra Cricket Association Stadium',inplace=True)
df_test1.replace('M Chinnaswamy Stadium','MA Chidambaram Stadium',inplace=True)
```

```python
venue = df_test1.iloc[0]['venue']
df123 = df.loc[df['venue']==venue,['match_id','ball','innings','batting_team','bowling_team','runs_off_bat','extras','player_dismissed']]

match_ids = df123['match_id'].unique()  
for id in match_ids:
    _df = df123.loc[df123['match_id']==id]
    df_temp=_df.loc[_df['ball']<6.1,['match_id','innings','batting_team','bowling_team','runs_off_bat','extras','player_dismissed']]
    df1=df_temp.loc[df_temp['innings'] == 1,['match_id','innings','batting_team','bowling_team','runs_off_bat','extras','player_dismissed'] ]
    df2=df_temp.loc[df_temp['innings'] == 2,['match_id','innings','batting_team','bowling_team','runs_off_bat','extras','player_dismissed'] ]

    # 'venue','innings','batting_team','bowling_team','runs_off_bat','extras','runs_ball'

    list_runs=['runs_off_bat','extras']
    df1["runs_ball"] = df1[list_runs].sum(axis=1).astype(int)
    runs_6overs_inning1= df1["runs_ball"].sum()
    #print(runs_6overs_inning1)
    df2["runs_ball"] = df2[list_runs].sum(axis=1).astype(int)
    runs_6overs_inning2= df2["runs_ball"].sum()
    #print(runs_6overs_inning2)

    wicket_inn1=df1.count()['player_dismissed']
    wicket_inn2=df2.count()['player_dismissed']
    
    batting_team = _df.iloc[0]['batting_team']
    bowling_team = _df.iloc[0]['bowling_team']

    X1= [[id,venue,1,batting_team,bowling_team,runs_6overs_inning1,wicket_inn1],[id,venue,2,bowling_team,batting_team,runs_6overs_inning2,wicket_inn2]]
    X = pd.DataFrame(X1, columns =['match_id','venue','innings','batting_team','bowling_team','runs_6overs','wickets'])
    main = pd.concat([main,X])
# print(main)
main.to_csv('file4.csv')
```

```python
df_main1 = pd.read_csv('file4.csv')

#One Hot encoding
df1 = df_main1.copy(deep = True)
onehotencoder = OneHotEncoder()

ven = onehotencoder.fit_transform(df1['venue'].values.reshape(-1,1)).toarray()
dfOneHot = pd.DataFrame(ven, columns = ["venue"+str(int(i)) for i in range(len(df1['venue'].unique()))])
df1 = pd.concat([df1, dfOneHot], axis=1)

bat = onehotencoder.fit_transform(df1['batting_team'].values.reshape(-1,1)).toarray()
dfOneHot = pd.DataFrame(bat, columns = ["bat"+str(int(i)) for i in range(len(df1['batting_team'].unique()))])
df1 = pd.concat([df1, dfOneHot], axis=1)

bowl = onehotencoder.fit_transform(df1['bowling_team'].values.reshape(-1,1)).toarray()
dfOneHot = pd.DataFrame(bowl, columns = ["bowl"+str(int(i)) for i in range(len(df1['bowling_team'].unique()))])
df1 = pd.concat([df1, dfOneHot], axis=1)

inn = onehotencoder.fit_transform(df1['innings'].values.reshape(-1,1)).toarray()
dfOneHot = pd.DataFrame(inn, columns = ["inn"+str(int(i)) for i in range(len(df1['innings'].unique()))])
df1 = pd.concat([df1, dfOneHot], axis=1)

df1.drop(['innings','venue','batting_team','bowling_team'], axis = 1,inplace = True)

X_main = df1.iloc[:,1:].values
Y_main = df1.iloc[:,0].values

#Splitting data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_main, Y_main, test_size = 0.25, random_state = 0)

#Using RandomForesrRegressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100,max_features=None)
rf.fit(X_train,y_train)

y_pred1 = rf.predict(X_test) 
# print('Random Forrest')
# print(sklearn.metrics.mean_absolute_error(y_test, y_pred1))
# print(sklearn.metrics.mean_squared_error(y_test, y_pred1))
# print(np.sqrt(sklearn.metrics.mean_squared_error(y_test, y_pred1)))
# print(y_test,y_pred1)
```
